## Hugging Face (HF):

Think of Hugging Face as a full ecosystem for building, sharing, and running AI models, especially in NLP, vision, audio, and multimodal AI. It’s not just a library—it’s closer to a “GitHub + runtime + tooling stack” for machine learning.

### Key libraries


#### [Datasets](https://huggingface.co/docs/datasets/index)
is an open-source Python library that provides easy access to large-scale datasets for machine learning and natural language processing. 
The library offers a unified framework to load, preprocess, and iterate over datasets from the Hugging Face Hub or local files. It supports datasets of any size, using efficient on-disk memory mapping via Apache Arrow. Its design allows data streaming, version control, and collaborative sharing across users and organizations.


#### [Tokenizers](https://huggingface.co/docs/tokenizers/index)
is an open-source library that provides fast, efficient text tokenization for natural language processing (NLP). 
It is notable for its high performance, 
written in Rust with Python bindings, and serves as a core component of the Transformers ecosystem.
It supports multiple tokenization algorithms—like 
**Byte Pair Encoding (BPE)**, 
**WordPiece**, and 
**SentencePiece**—and allows efficient handling of large text corpora. 
Its design aims to combine the speed of compiled languages with the flexibility of Python APIs.

#### [Transformers](https://huggingface.co/docs/transformers/en/index)
is an open-source Python library for working with transformer-based machine learning models. 
It offers a unified API for thousands of pre-trained models, enabling developers to perform tasks such as sentiment analysis, translation, and text summarization with minimal setup. It abstracts complex model architectures—like BERT, GPT, and T5—into simple interfaces, reducing the need for custom implementations.


#### [Accelerate](https://huggingface.co/docs/accelerate/en/index)
Accelerate is an open-source Python library that simplifies the process of training and deploying machine learning models on different hardware setups. 
It provides a unified interface for scaling code across CPUs, GPUs, and TPUs without rewriting model or optimizer logic. 
It is especially relevant for deep learning practitioners using PyTorch or TensorFlow.

Accelerate abstracts away hardware-specific code by automatically managing distributed training and mixed precision. 
It allows users to write plain PyTorch code and scale seamlessly from single-GPU laptops to multi-node clusters. With minimal configuration, it can synchronize gradients, manage device placement, and optimize performance for available resources.

#### [Parameter-Efficient Fine-Tuning (PEFT)](https://huggingface.co/docs/peft/en/index)
is an open-source library that enables efficient fine-tuning of large pre-trained models by modifying only a small subset of parameters. 
It allows users to adapt massive models such as GPT, BERT, or Stable Diffusion for new tasks while dramatically reducing computational and memory costs compared with full-model fine-tuning.

The PEFT library implements state-of-the-art methods that fine-tune large language or vision models by introducing lightweight, trainable modules—such as low-rank adapters—while freezing the bulk of model weights. This approach yields models that achieve accuracy comparable to full fine-tuning but require only a fraction of GPU memory and storage, making model customization accessible even on consumer hardware.

#### Accelerate:

```python
from accelerate import Accelerator


accelerator = Accelerator()
model = torch.nn.Linear(10, 2)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

model, optimizer = accelerator.prepare(model, optimizer)

x = torch.randn(4, 10, device=device)
y = model(x)
loss = y.mean()
accelerator.backward(loss)

optimizer.step()
```

#### PEFT:

```python
# complet code snippet using a toy example:

from datasets import Dataset
from transformers import(
    AutoModelForCausalLM,
    AutoTokenizer,
    DataCollatorForLanguageModeling, 
    Trainer, 
    TrainingArguments,
)

from peft import LoraConfig, get_peft_model, TaskType

# ----------------------------
# 1) Tiny toy dataset
# ----------------------------
train_texts = [
    "Question: What is Hugging Face?\nAnswer: Hugging Face is a company and open-source ecosystem for machine learning.",
    "Question: What is PEFT?\nAnswer: PEFT stands for Parameter-Efficient Fine-Tuning.",
    "Question: What is LoRA?\nAnswer: LoRA adapts a model by training small low-rank matrices instead of all weights.",
    "Question: Why use LoRA?\nAnswer: LoRA reduces memory usage and the number of trainable parameters.",
    "Question: What does a tokenizer do?\nAnswer: A tokenizer converts text into token IDs for a model.",
]

dataset = Dataset.from_dict({"text": train_texts})

# ----------------------------
# 2) Load tokenizer + base model
# ----------------------------

model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
# GPT-style tokenizers often need a pad token set for batching
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name)
model.config.pad_token_id = tokenizer.pad_token_id

# ----------------------------
# 3) Tokenize dataset
# ----------------------------
def tokenize_fn(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        max_length=128,
    )

tokenized_dataset = dataset.map(
    tokenize_fn, batched=True, 
    remove_columns=["text"], 
)

# For causal language modeling, HF docs use mlm=False
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,
)
# ----------------------------
# 4) Configure LoRA and wrap model with PEFT
# ----------------------------
# The PEFT docs show creating a LoraConfig and then calling get_peft_model().
# target_modules can vary by architecture. For GPT-2 / DistilGPT2, "c_attn" is a common choice

peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8, 
    lora_alpha=16,
    lora_dropout=0.1,
    target_modules=["c_attn"],
    bias="none",
)

model = get_peft_model(model, peft_config)

# Prints how many parameters will actually train
model.print_trainable_parameters()


# ----------------------------
# 5) Training setup
# ----------------------------
training_args = TrainingArguments(
    output_dir="./results/peft-lora-distilgpt2",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=1,
    learning_rate=2e-4,
    logging_steps=1,
    save_strategy="epoch",
    eval_strategy="no",
    report_to="none",
    fp16=False,   # set True on compatible CUDA GPUs
    bf16=False,   # set True on compatible hardware
)

trainer = Trainer(
    model=model, 
    args=training_args,
    train_dataset=tokenized_dataset, 
    data_collator=data_collator,
)

# ----------------------------
# 6) Train
# ----------------------------
trainer.train()


# ----------------------------
# 7) Save adapter + tokenizer
# ----------------------------
save_dir = "./results/peft-lora-distilgpt2/final"
model.save_pretrained(save_dir)
tokenizer.save_pretrained(save_dir)

print(f"Saved PEFT Adapter + Tokenizer to {save_dir}")



```

```python
# ----------------------------
# 8) Run inference with the trained adapter
# ----------------------------

prompt = "Question: What is PEFT and LORA?\nAnswer:"
inputs = tokenizer(prompt, return_tensors="pt")


# If you're on GPU, move inputs/model to CUDA:
# model = model.to("cuda")
# inputs = {k: v.to("cuda") for k, v in inputs.items()}

model = model.to(device)
inputs = {k:v.to(device) for k,v in inputs.items()}

generated = model.generate(
    **inputs,
    max_new_tokens=20,
    do_sample=True,
    top_k=10,
    top_p=0.95,
    temperature=0.7,
)

print(
    tokenizer.decode(generated[0], skip_special_tokens=True)
    )
```

The above model/training setup is too weak for the task, i.e., more epochs, larger datasets, and various other matters must be considered.
Nevertheless the following code snippet can improve the quality of answer due to the following facts:

- model.eval() disables training behavior like dropout
-torch.no_grad() avoids unnecessary gradient tracking
- do_sample=False makes output deterministic and less random


```python
model = model.to(device)
model.eval()

inputs = tokenizer(prompt, return_tensors="pt")
inputs = {k: v.to(device) for k, v in inputs.items()}

with torch.no_grad():
    generated = model.generate(
        **inputs,
        max_new_tokens=20,
        do_sample=False,
        pad_token_id=tokenizer.eos_token_id,
    )

print(tokenizer.decode(generated[0], skip_special_tokens=True))
```
